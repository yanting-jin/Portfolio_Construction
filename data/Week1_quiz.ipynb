```python
def get_EW_returns():
    """
    Load and format the EDHEC Hege Fund Index Returns
    """
    EW = pd.read_csv('/Users/Jinyanting/Documents/GitHub/PortfolioConstruction/Week1/data/Portfolios_Formed_on_ME_monthly_EW.csv',
                     header=0, index_col=0, parse_dates=True,na_values = -99.99)
    EW = EW/100
    EW.index = pd.to_datetime(EW.index,format= '%Y%m').to_period('M')
    return EW
```
