# Construct the Market Cap Index

```python
ind_mktcap = ind_nfirms*ind_size
ind_mktcap.head()
```

```python
#total_mktcap = ind_mktcap.sum(axis = 'columns') # sum across all columns, or axis = 1
total_mktcap = ind_mktcap.sum(axis = 1) 
total_mktcap.plot()
```

```python
#capitalization weight
#ind_capweight = ind_mktcap.divide(total_mktcap, axis = 'rows')

ind_capweight = ind_mktcap.divide(total_mktcap, axis = 0)
```

```python
total_market_return = (ind_capweight*ind_returns).sum(axis = 'columns')
total_market_return.plot()
```

```python
total_market_index=erk.drawdown(total_market_return).Wealth  #  wealth_index = 1000*(1+returns_series).cumprod()
total_market_index.plot(title='total_market_index',label = 'total_market_index',legend=True)
total_mktcap.plot(label='total_mktcap',legend = True, secondary_y=True)
```

```python
# compute the 36 month rolling mean return, that is the moving average of the time series
total_market_index['1980':].plot(figsize=(12,6))
total_market_index['1980':].rolling(window=36).mean().plot()
```

```python
# total market index- trailing 36 months returns
tmi_tr36rets = total_market_return.rolling(window=36).aggregate(erk.annualize_rets,period_per_year = 12)
tmi_tr36rets.plot(label='Tr 36 mo rets',figsize = (12,5),legend=True)
total_market_return.plot(label='Returns',legend=True)
```

```python
tmi_tr36rets.plot(label='Tr 36 mo rets',figsize = (12,5),legend=True)
```

### Rolling Correlation - along with Multi-Indexes and .groupby

```python
# Multi-index correlation matrix with time and sectors
ts_corr = ind_returns.rolling(window=36).corr()
ts_corr.tail()
```

```python
# at the given date, all the industry correlation matrix
ind_tr36corr = ts_corr.groupby(level='date').apply(lambda cormat: cormat.values.mean()) # for each row, return the mean value of the correlation value
ind_tr36corr.plot()
```

```python
ind_tr36corr.tail()
```

```python
tmi_tr36rets.plot(label='Tr36 Mo Rets', figsize=(12,6),legend=True)
ind_tr36corr.plot(label='Tr36 Mo Corr', legend=True,secondary_y=True)
```

When the market is pluging , the correlation increases and vice versa. 
Thus, diversification still suffers from systematic risks

This is easy to understand intuitively as the market increases, all stocks across sectors increase; and the reverse is also true.
